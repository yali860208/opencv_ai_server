```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show_rgb_img(img):
    img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.show()
```

```python
def modify_intensity(file):
    img = cv2.imread(file)
    origin_img = img
    print("origin picture:")
    show_rgb_img(origin_img)

    maxIntensity = 255.0 # depends on dtype of image data

    # Parameters for manipulating image data
    phi = 1
    theta = 1.06

    # Increase intensity 
    increase_img = (maxIntensity/phi)*(origin_img/(maxIntensity/theta))**0.5
    increase_img = np.array(increase_img, dtype=np.uint8)
    print("Increase intensity :")
    show_rgb_img(increase_img)

    # Decrease intensity  
    decrease_img = (maxIntensity/phi)*(origin_img/(maxIntensity/theta))**2
    decrease_img = np.array(decrease_img, dtype=np.uint8)
    print("Decrease intensity :")
    show_rgb_img(decrease_img)
```

```python
modify_intensity("onew.jpg")
```

```python
# HLS : Hue(顏色)、Lightness(亮度)、Saturation(飽和度)
def modify_lightness_saturation(file,lightness=0,saturation=0):
    img = cv2.imread(file)
    origin_img = img
    fImg = img.astype(np.float32)
    fImg = fImg / 255.0
    hlsImg = cv2.cvtColor(fImg, cv2.COLOR_BGR2HLS)
    hlsCopy = np.copy(hlsImg)
    
    hlsCopy[:,:,1] = (1 + lightness/100.0) * hlsCopy[:,:,1]
    hlsCopy[:,:,1][hlsCopy[:,:,1]>1] = 1
    
    hlsCopy[:,:,2] = (1 + saturation/100.0) * hlsCopy[:,:,2]
    hlsCopy[:,:,2][hlsCopy[:,:,2]>1] = 1
    
    result_img = cv2.cvtColor(hlsCopy, cv2.COLOR_HLS2BGR)
    result_img = ((result_img*255).astype(np.uint8))
    print("origin picture:")
    show_rgb_img(origin_img)
    print("after picture:")
    show_rgb_img(result_img)
```

```python
modify_lightness_saturation("onew.jpg",0,120)
```
